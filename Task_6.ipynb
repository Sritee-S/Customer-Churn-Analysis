```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('Telco_Customer_Churn_Dataset.csv')

# Clean 'TotalCharges'
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'].replace(" ", pd.NA))
df.dropna(subset=['TotalCharges'], inplace=True)

# Convert target to binary for easier plotting
df['Churn_binary'] = df['Churn'].map({'No':0, 'Yes':1})

# Set seaborn style
sns.set(style="whitegrid")
```

```python
# 1. Box plot: MonthlyCharges by Churn
plt.figure(figsize=(8,6))
sns.boxplot(x='Churn', y='MonthlyCharges', hue='Churn' , data=df, palette='Set2')
plt.title('Box Plot of Monthly Charges by Churn Status')
plt.show()
```

```python
# 2. Violin plot: Tenure by Churn
plt.figure(figsize=(8,6))
sns.violinplot(x='Churn', y='tenure', hue='Churn' , data=df, palette='Set1')
plt.title('Violin Plot of Tenure by Churn Status')
plt.show()
```

```python
# 3. Pair plot of selected numeric variables colored by Churn
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
sns.pairplot(df[numeric_cols + ['Churn']], hue='Churn', palette='coolwarm')
plt.suptitle('Pair Plot of Tenure, Monthly Charges, and Total Charges by Churn', y=1.02)
plt.show()
```

```python
# 4. Count plot for Contract type vs Churn
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Contract', hue='Churn', palette='pastel')
plt.title('Customer Count by Contract Type and Churn')
plt.show()
```

```python
# 5. Strip plot (scatter) of MonthlyCharges by Contract type colored by Churn
plt.figure(figsize=(10,6))
sns.stripplot(data=df, x='Contract', y='MonthlyCharges', hue='Churn', jitter=True, dodge=True, palette='Set1', alpha=0.7)
plt.title('Monthly Charges Distribution by Contract Type and Churn')
plt.show()
```
